Autenticacao

```python
url_base_api = 'https://api.github.com'
owner = 'amzn'
url = url_base_api+'/users/'+owner+'/repos'
url
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
acess_token = os.getenv('acess_token')

headers = {
    'X-GitHub-Api-Version': '2022-11-28',
    'Authorization' :'Bearer '+ acess_token
    }

```

Extracao de dados

```python
repos_list = []
for page_num in range(1,6):
    try:
        url_page = url + f'?page={page_num}'
        response = requests.get(url_page,headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
qtd_paginas = len(repos_list)
```

```python
qtd_paginas
```

Tratamento de dados

```python
repos_list[0][2]['name']
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name']) 
len(repos_name)
```

```python
repos_name[:10]
```

```python
repos_list[1][1]['language']
```

```python
repos_language = []
for page in repos_list:
    for repo in page:
        repos_language.append(repo['language'])
len(repos_language)
```

```python
repos_language[:10]
```

```python
import pandas as pd
```

```python
dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name
dados_amz['repository_language'] = repos_language
```

```python
dados_amz
```

```python
dados_amz.to_csv('../data_processed/amazon.csv')
```

Alterando formato do arquivo, para conseguir alocar ele no armazenamento de dados que nesse caso Ã© um repositorio do github

```python
with open('../data_processed/amazon.csv','rb') as file:
    file_content = file.read()

encoded_content = base64.b64encode(file_content)
    
```

upload para o armazenamento de dados

```python
import os
from dotenv import load_dotenv
load_dotenv()


url_base_api = 'https://api.github.com'
username = os.getenv('owner_git')
repo = 'linguagens-utilizadas'
path ='amazon.csv'

url = f'{url_base_api}/repos/{username}/{repo}/contents/{path}'
url
```
